```python
score = []
for i in range(10):
    score.append(play_game(oneturnheuristic(),oneturnai(),board,True))
    wins = [1 for w in score if w == 1]
    looses = [1 for l in score if l == -1]
    draws = [1 for d in score if d == 0]
    wins = sum(wins)
    looses = sum(looses)
    draws = sum(draws)
print 'When the heuristic AI starts first, it wins: ' + str(wins) + ', looses: ' + str(looses) + ', draws: ' + str(draws) + ' out of 1000 games with the normal One Turn Ahead AI'
print wins + looses + draws
```
